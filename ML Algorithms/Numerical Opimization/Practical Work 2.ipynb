## stochastic

```python
from sklearn.metrics import r2_score
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,20)
y=-2*x+1
theta_0=0
theta_1=0
alpha=0.001
max_iter=50
h_x_list=[]
cost_fn_list=[]
theta_0_list=[]
theta_1_list=[]
for i in range (max_iter):
    for j in range(len(x)):
        h_x=theta_0+theta_1*x[j]
        cost=0.5*(h_x-y[j])
        cost_fn_list.append(cost)
        grad_theta0 = (h_x-y[j])
        grad_theta1 = (h_x-y[j])*x[j]
        theta_0 = theta_0 - alpha * grad_theta0
        theta_0_list.append(theta_0)
        theta_1 = theta_1 - alpha * grad_theta1
        theta_1_list.append(theta_1)
h_x=theta_0+theta_1*x
plt.subplot(2, 2, 1)
plt.scatter(x,y)
plt.plot(x,h_x,'r')
plt.xlabel("x")
plt.ylabel("y")

plt.subplot(2, 2, 2)
plt.plot(cost_fn_list)
plt.xlabel("iteration")
plt.ylabel("cost function")
plt.subplot(2, 2, 3)
plt.plot(theta_0_list,cost_fn_list)
plt.xlabel("theta_0")
plt.ylabel("cost function")
plt.subplot(2, 2, 4)
plt.plot(theta_1_list,cost_fn_list)
plt.xlabel("theta_1")
plt.ylabel("cost function")
plt.show()
print(r2_score(y,h_x))
```

# stochastic with  stopping criteria

```python
from sklearn.metrics import r2_score
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,20)
y=-2*x+1
theta_0=0
theta_1=0
alpha=0.001
max_iter=1000
h_x_list=[]
cost_fn_list=[]
theta_0_list=[]
theta_1_list=[]

for i in range(max_iter):
        for j in range(len(x)):
           h_x = theta_0 + theta_1 * x[j]
           cost = 0.5 * (h_x - y[j])
           cost_fn_list.append(cost)
           grad_theta0 = (h_x - y[j])
           grad_theta1 = (h_x - y[j]) * x[j]
           theta_0 = theta_0 - alpha * grad_theta0
           theta_0_list.append(theta_0)
           theta_1 = theta_1 - alpha * grad_theta1
           theta_1_list.append(theta_1)
        if ((grad_theta0) ** 2 + (grad_theta1) ** 2) ** 0.5 < 0.001:
            print(i)
            break

h_x=theta_0+theta_1*x
plt.subplot(2, 2, 1)
plt.scatter(x,y)
plt.plot(x,h_x)
plt.xlabel("x")
plt.ylabel("y")

plt.subplot(2, 2, 2)
plt.plot(cost_fn_list)
plt.xlabel("iteration")
plt.ylabel("cost function")
plt.subplot(2, 2, 3)
plt.plot(theta_0_list,cost_fn_list)
plt.xlabel("theta_0")
plt.ylabel("cost function")
plt.subplot(2, 2, 4)
plt.plot(theta_1_list,cost_fn_list)
plt.xlabel("theta_1")
plt.ylabel("cost function")
plt.show()
print(r2_score(y,h_x))
```

## mini-batch

```python
from sklearn.metrics import r2_score
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,20)
y=-2*x+1
n = 10
theta_0=0
theta_1=0
alpha=0.001
max_iter=1000
h_x_list=[]
cost_fn_list=[]
theta_0_list=[]
theta_1_list=[]
for i in range (max_iter):
    for j in range(n):
        h_x = theta_0 + theta_1 * x[i*n:n*(i+1)]
        h_x_list.append(h_x)
        cost_fn = (1 / 2 * n) * sum(h_x - y[i*n:n*(i+1)]) ** 2
        cost_fn_list.append(cost_fn)
        grad_theta0 = (1 / n) * (sum(h_x - y[i*n:n*(i+1)]))
        grad_theta1 = (1 / n) * (sum((h_x - y[i*n:n*(i+1)]) * x[i*n:n*(i+1)]))
        theta_0 = theta_0 - alpha* grad_theta0
        theta_0_list.append(theta_0)
        theta_1 = theta_1 - alpha * grad_theta1
        theta_1_list.append(theta_1)
h_x=theta_0+theta_1*x
plt.subplot(2, 2, 1)
plt.scatter(x,y)
plt.plot(x,h_x,'r')
plt.xlabel("x")
plt.ylabel("y")

plt.subplot(2, 2, 2)
plt.plot(cost_fn_list)
plt.xlabel("iteration")
plt.ylabel("cost function")
plt.subplot(2, 2, 3)
plt.plot(theta_0_list,cost_fn_list)
plt.xlabel("theta_0")
plt.ylabel("cost function")
plt.subplot(2, 2, 4)
plt.plot(theta_1_list,cost_fn_list)
plt.xlabel("theta_1")
plt.ylabel("cost function")
plt.show()
print(r2_score(y,h_x))
```
