# LR Full Implementation

```python
path="E:\RegData.csv"
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
RegData = np.genfromtxt(path,delimiter=',')
x=RegData[:,0]
y=RegData[:,1]


theta_0=0
theta_1=0
m = len(x)
alpha=0.0001
max_iter=1000
h_x_list=[]
cost_fn_list=[]
theta_0_list=[]
theta_1_list=[]
for i in range(max_iter):
        h_x = theta_0 + theta_1 * x
        h_x_list.append(h_x)
        cost_fn = (1 / 2 * m) * sum(h_x - y) ** 2
        cost_fn_list.append(cost_fn)
        grad_theta0 = (1 / m) * (sum(h_x - y))
        grad_theta1 = (1 / m) * (sum((h_x - y) * x))
        theta_0 = theta_0 - alpha* grad_theta0
        theta_0_list.append(theta_0)
        theta_1 = theta_1 - alpha * grad_theta1
        theta_1_list.append(theta_1)

for i in range(len(h_x_list)):
    line=h_x_list[i]
    plt.plot(x,line)

plt.show()
plt.subplot(2, 2, 1)
plt.scatter(x,y)
plt.plot(x,h_x)
plt.xlabel("x")
plt.ylabel("y")
plt.subplot(2, 2, 2)
plt.plot(cost_fn_list)
plt.xlabel("iteration")
plt.ylabel("cost function")
plt.subplot(2, 2, 3)
plt.plot(theta_0_list,cost_fn_list)
plt.xlabel("theta_0")
plt.ylabel("cost function")
plt.subplot(2, 2, 4)
plt.plot(theta_1_list,cost_fn_list)
plt.xlabel("theta_1")
plt.ylabel("cost function")
plt.show()
print(r2_score(y,h_x))
##to enhance r2 score improve alpha
```
