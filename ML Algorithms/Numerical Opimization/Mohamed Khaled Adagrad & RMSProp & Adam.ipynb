```python

from sklearn.metrics import r2_score
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,20)
y=-1*x+2
theta_0=0
theta_1=0
v_theta_0=0
v_theta_1=0
alpha=0.01
epsilon=1e-8
beta=0.9
max_iter=1000
h_x_list=[]
cost_fn_list=[]
theta_0_list=[]
theta_1_list=[]
m=len(x)
for i in range(max_iter):
        h_x = theta_0 + theta_1 * x
        h_x_list.append(h_x)
        cost_fn = (1 / (2 * m)) * sum(h_x - y) ** 2
        cost_fn_list.append(cost_fn)
        grad_theta0 = (1 / m) * (sum(h_x - y))
        grad_theta1 = (1 / m) * (sum((h_x - y) * x))
        v_theta_0=beta*v_theta_0+(1-beta)*grad_theta0**2
        v_theta_1 = beta*v_theta_1 +(1-beta)*grad_theta1**2
        theta_0 = theta_0 - (alpha/(v_theta_0**0.5+epsilon))*grad_theta0
        theta_0_list.append(theta_0)
        theta_1 = theta_1 - (alpha/(v_theta_1**0.5+epsilon))*grad_theta1
        theta_1_list.append(theta_1)
        # if ((grad_theta0) ** 2 + (grad_theta1) ** 2) ** 0.5 < 0.001:
        if i > 1 and (cost_fn_list[i - 1] - cost_fn_list[i] < 0.001):
            print("the # of iteration =",i)
            break

h_x=theta_0+theta_1*x
plt.subplot(2, 2, 1)
plt.scatter(x,y)
plt.plot(x,h_x,"r")
plt.xlabel("x")
plt.ylabel("y")

plt.subplot(2, 2, 2)
plt.plot(cost_fn_list)
plt.xlabel("iteration")
plt.ylabel("cost function")
plt.subplot(2, 2, 3)
plt.plot(theta_0_list,cost_fn_list)
plt.xlabel("theta_0")
plt.ylabel("cost function")
plt.subplot(2, 2, 4)
plt.plot(theta_1_list,cost_fn_list)
plt.xlabel("theta_1")
plt.ylabel("cost function")
plt.show()
print(r2_score(y,h_x))
```
